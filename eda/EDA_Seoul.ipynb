#### 중구에서 ‘5층 이하’ 주택의 평균 거래가격이 ‘10층 이하’보다 높게 나타난 이유

```python
# 1) 단독다가구 비율 계산 함수
def get_dandok_ratio(df, gu_name):
    ct = df['건물용도'].value_counts(normalize=True) * 100
    dandok = ct.get('단독다가구', 0)
    return dandok

# 2) 세 구별 단독다가구 비율 계산
ratios = {
    '강서구': get_dandok_ratio(강서, '강서구'),
    '중구': get_dandok_ratio(중구, '중구'),
    '송파구': get_dandok_ratio(송파, '송파구')
}

# 3) 막대그래프로 비교
plt.figure(figsize=(6,4))
plt.bar(ratios.keys(), ratios.values(), color=['#7E57C2','#42A5F5','#66BB6A'])
plt.title('강서구·중구·송파구 단독다가구 비중(%)')
plt.ylabel('단독다가구 비중(%)')

# y축 범위 자동 조정 (최댓값보다 20~30% 크게)
ymax = max(ratios.values()) * 1.3
plt.ylim(0, ymax)

# 그래프 위에 숫자 표시
for i, v in enumerate(ratios.values()):
    plt.text(i, v+0.5, f'{v:.1f}%', ha='center', va='bottom')

plt.show()

# 4) 표 형태로도 출력
import pandas as pd
pd.DataFrame.from_dict(ratios, orient='index', columns=['단독다가구 비중(%)'])
```

결정적 요인: **단독다가구 비중**
- 단독다가구는 평균 거래가격이 가장 높은 유형이며, 주로 5층 이하 구간에 속한다. (이는 3-1. 건물용도와 물건금액의 상관관계 분석에서 확인된 바 있다.)
- 중구는 5층 이하에서 단독다가구 비중이 타 구 대비 높게 분포한다.
- 결과적으로, 중구의 5층 이하 평균가격이 10층 이하보다 높게 역전되는 구성효과가 발생한다.

```python
# 구별 데이터
gus = {'강서구': 강서, '중구': 중구, '송파구': 송파}

# 라벨 순서 & 색상 고정
use_order = ['단독다가구', '아파트', '연립다세대', '오피스텔']
color_map = {
    '단독다가구': '#D6C8FF',
    '아파트':     '#FFB7B2',
    '연립다세대': '#AEE6E6',
    '오피스텔':   '#F9E79F'
}

# 퍼센트 라벨 (작은 조각은 표시 X)
def autopct_threshold(values, min_pct=3.0):
    total = sum(values)
    def _fmt(pct):
        return f'{pct:.1f}%' if pct >= min_pct else ''
    return _fmt

# 2행 3열 (강서/중구/송파 × 40/50평대)
fig, axes = plt.subplots(2, 3, figsize=(15, 8), facecolor='#F9FCE8', constrained_layout=True)
targets = ['40평대', '50평대']

for col, (gu_name, df_gu) in enumerate(gus.items()):
    for row, size_label in enumerate(targets):
        ax = axes[row, col]
        sub = df_gu[df_gu['건물면적구분'] == size_label]
        counts = sub['건물용도'].value_counts().reindex(use_order, fill_value=0)

        vals = counts.values
        labels = counts.index.tolist()
        colors = [color_map[u] for u in labels]

        if vals.sum() == 0:
            ax.axis('off')
            ax.set_title(f'{gu_name} {size_label}\n(데이터 없음)', fontsize=13)
            continue

        # 파이차트 (항목 라벨은 제거, 퍼센트만 표시)
        wedges, texts, autotexts = ax.pie(
            vals,
            labels=None,                               # 조각 위 항목 라벨 제거
            autopct=autopct_threshold(vals, 3.0),      # 3% 미만 퍼센트 숨김
            startangle=90,
            colors=colors,
            textprops={'fontsize': 9, 'fontweight': 'bold'}
        )
        ax.set_aspect('equal')
        ax.set_title(f'{gu_name} — {size_label}', fontsize=14, fontweight=('bold' if row == 0 else 'normal'))

# 공통 범례
handles = [plt.Line2D([0],[0], marker='o', color='w', label=u,
                      markerfacecolor=color_map[u], markersize=10)
           for u in use_order]
fig.legend(handles=handles, labels=use_order, title='건물용도',
           loc='center right', bbox_to_anchor=(1.02, 0.5))

plt.suptitle(
    '40평대 / 50평대 건물용도 비율 비교 (강서·중구·송파)',
    fontsize=16,
    fontweight='bold',
    y=1.05   # 👈 y축 위치를 0.98 → 1.05 로 올림
)

plt.show()
```

단독다가구의 평균 가격은 높게 형성되어있음에도 불구하고, 아파트의 비율이 높은 40평대가 50평대보다 평균 가격이 높게 형성된 이유는 무엇일까?

```python
# 데이터 준비 (강서구 예시, 중구/송파도 동일하게 가능)
def plot_usage_ratio(df, gu_name):
    # 교차표 (평수대 × 건물용도 → 건수)
    crosstab = pd.crosstab(df['건물면적구분'], df['건물용도'])
    share = crosstab.div(crosstab.sum(axis=1), axis=0) * 100
    
    # 색상 고정
    color_map = {
        '단독다가구': '#D6C8FF',
        '아파트': '#FFB7B2',
        '연립다세대': '#AEE6E6',
        '오피스텔': '#F9E79F'
    }
    use_order = ['단독다가구','아파트','연립다세대','오피스텔']
    share = share.reindex(columns=use_order, fill_value=0)

    # 그래프
    share.plot(kind='bar', stacked=True, figsize=(10,6),
               color=[color_map[u] for u in use_order])
    
    plt.title(f"{gu_name} 평수대별 건물용도 비율 (가로=100%)", fontsize=14, fontweight='bold')
    plt.xlabel("건물면적구분"); plt.ylabel("비율(%)")
    plt.legend(title="건물용도", bbox_to_anchor=(1.02, 1), loc='upper left')
    plt.ylim(0,100)
    plt.tight_layout()
    plt.show()

# 실행 예시
plot_usage_ratio(강서, "강서구")
plot_usage_ratio(중구, "중구")
plot_usage_ratio(송파, "송파")
```

1. 단독다가구의 평균 가격 수준

- 세 구 모두에서 단독다가구의 평균 거래가격은 아파트, 연립다세대, 오피스텔 대비 높게 나타났다.

- 그러나 단독다가구의 비중은 대체로 60평대 이상 대형 평수에서 집중적으로 나타나며, 중·소형 평수에서는 상대적으로 낮은 수준에 머무른다.

2. 아파트 비율의 차이

- 강서구와 송파구에서는 40평대에서 아파트 비중이 70~90% 수준으로 매우 높다.

- 반면 50평대에서는 아파트 비중이 줄고, 단독다가구 비율이 크게 상승한다.

3. 평수대별 가격 역전 현상

- 강서구와 송파구에서 관찰되는 “40평대 > 50평대” 평균가격 역전 현상은,

    * 단독다가구의 가격 자체가 높음에도 불구하고,

    * 40평대에서는 ‘비싼 아파트’의 비중이 높고

    * 50평대에서는 단독다가구 비중이 늘어나면서 아파트 비중이 낮아지는 구성효과에 기인한다.
